# Model
In diesem Notebook wird eine Pipeline erstellt, mit dieser soll der Datensatz automatisiert auf verschiedene Modelle trainiert und evaluiert werden.

Der DP-WGAN lieferte die besten Ergebnisse, wenn er auf ein Standard skaliertes Datenset trainiert wurde.
Die generierten synthetischen Daten sind aus diesem Grund ebenfalls Standard skaliert.

Aus diesem Grund wird in diesem Notebook hauptsächlich auf Standard skalierten Daten gearbeitet.
Somit sind df_train und df_test die realen, skalierten Daten.

#### Utils

```python
def draw_auc_roc_curve(y_test, y_pred):  
    fpr,tpr,_ = metrics.roc_curve(y_test,y_pred)
    rocAuc = metrics.auc(fpr, tpr)
    plt.figure(figsize=(12,6))
    plt.title('ROC Curve')
    sns.lineplot(fpr, tpr, label = 'AUC for Model = %0.2f' % rocAuc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()

def xy_split(table:pd.DataFrame):
    """Takes a full Give Me some Credit Dataset
        Returns X, y splitted Dataset
    """
    table.drop("Unnamed: 0", axis=1, errors="ignore", inplace=True)
    y = table["SeriousDlqin2yrs"]
    X = table.drop("SeriousDlqin2yrs", axis=1)
    return X, y

def scaleX(table_X: pd.DataFrame) -> pd.DataFrame:
    """StandardScales the X values of a DataFrame.
    Do not include Targetvariable, as it would also Scale it
    """
    scaler = preprocessing.StandardScaler()
    scaled = scaler.fit_transform(table_X)
    return pd.DataFrame(data=scaled, columns=table_X.columns).copy()

def scaleTable(table: pd.DataFrame) -> pd.DataFrame:
    """This Function StandardScales a whole Dataset.
    It should also contain Targetvariable
    """
    X, y = xy_split(table)
    X = scaleX(X)
    return pd.concat([y, X], axis=1)

def calc_auc_real(model, draw_curve=True):
    proba = model.predict_proba(df_test_X)[:, 1]
    if draw_curve:
        draw_auc_roc_curve(df_test_y, proba)
    return roc_auc_score(df_test_y, proba)

def calc_f1_real(model):
    proba = model.predict(df_test_X)
    return f1_score(df_test_y, proba)
```

## Build Evaluation Pipeline

```python
def train_pipeline(X, y):
    print("Start Training")
    
    # Run Models
    scores = []
    for model_name, mp in model_params.items():
        # Test GridSearchCV
        clf =  GridSearchCV(mp['model'], mp['params'], cv=5, scoring="roc_auc", verbose=1)
        clf.fit(X, y)
        scores.append({
            'model': model_name,
            'best_score': clf.best_score_,
            'best_params': clf.best_params_,
            "clf": clf
        })
        print(f"{model_name}: {clf.best_score_}, {clf.best_params_}")
    df = pd.DataFrame(scores,columns=['model','best_score','best_params', "clf"])
    return df
```

```python
def evaluate_pipeline(table, scaled=True):
    """The Downstream Metrics, are evaluated on scaled Dataset, so they are only available if the input data is StandardScaled.
    """
    X, y = xy_split(table)

    print("Data Splitted")
    result = train_pipeline(X, y)
    
    best_model = result["clf"][result["best_score"].idxmax()]
    
    if scaled:
        # Die Metriken werden auf ein Standard skaliertes Testset getestet
        plot_confusion_matrix(best_model, df_test_X, df_test_y)
        print(f"AUC Score: {calc_auc_real(best_model)}")
        print(f"F1 Score: {calc_f1_real(best_model)}")

    return result
```
